```python
liste = []
somme =  0
produit = 1
moyenne = 0

for i in range(0, 10) :
    nbre = int(input(f"Veuillez saisir un nombre{i} : "))
    liste.append(nbre)

for j in liste :
    somme += j
    produit *= j
moyenne = somme / len(liste)

print(f"La somme des nombres est {somme}, le produit des nombres est {produit} et la moyenne des nombres est {moyenne}")
```
