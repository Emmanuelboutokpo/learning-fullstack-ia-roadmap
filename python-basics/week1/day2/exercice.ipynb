```python
# Exercice 1 : Voir si un nombre est pair ou impair.
# Objectif : utiliser les entrées utilisateur, les opérations arithmétiques et les structures conditionnelles.

nombre = int(input('Veuillez entrer un nombre : '))

if nombre % 2 == 0 :
    print(f"Le nombre {nombre} est pair ! ")
else : 
    print(f"Le nombre {nombre} est impaire ! ")
```

```python
#Exercice 5: Calcul de la puissance de deux nombres réels
#Objectif : utiliser les entrées utilisateur et les opérations arithmétiques.
nombre_x = int(input('Veuillez saisir X : '))
nombre_y = int(input('Veuillez saisir Y : '))

puissance_X_Y = nombre_x**nombre_y

print(f"La puissance de {nombre_x} et {nombre_y} est {puissance_X_Y} !")
```

```python
# Exercice 6 : calcule de la somme, du produit, la soustraction et la divison de deux nombres
# Objectif : utiliser les entrées utilisateur et les opérations arithmétiques.

nombre_x = int(input('Veuillez saisir X : '))
nombre_y = int(input('Veuillez saisir Y : '))

somme = nombre_x + nombre_y 
produit = nombre_x * nombre_y 
sub = nombre_x - nombre_y 
division = nombre_x / nombre_y 

print(f"La somme de {nombre_x} et {nombre_y} est {somme} !")
print(f"La produit de {nombre_x} et {nombre_y} est {produit} !")
print(f"La soustraction de {nombre_x} et {nombre_y} est {sub} !")
print(f"La division de {nombre_x} et {nombre_y} est {division} !")
```

```python
# Exercice 7 : Verification de signe de deux nombres
# Objectif : utiliser les entrées utilisateur et les structures conditionnelles.

nombre_x = int(input('Veuillez saisir X : '))
nombre_y = int(input('Veuillez saisir Y : '))

if (nombre_x > 0 and nombre_y > 0) or (nombre_x < 0 and nombre_y < 0): 
    print(f"Les nombres {nombre_x} et {nombre_y} ont mêmes signes !")
else : 
    print(f"Les nombres {nombre_x} et {nombre_y} n'ont pas mêmes signes !")
```

```python
# Exercice 8 : Echanger des valeurs numériques et verification
# Objectif : utiliser les entrées utilisateur, les opérations arithmétiques et les structures conditionnelles.
nombre_x = int(input('Veuillez saisir X : '))
nombre_y = int(input('Veuillez saisir Y : '))
if nombre_x*nombre_y > 0 : 
   temp = 0 
   temp = nombre_x
   nombre_x = nombre_y
   nombre_y = temp

   print(f"Voici l'echange : {nombre_x} et {nombre_y}")

else :
   nombre_x_som = nombre_x + nombre_y
   nombre_y_prod = nombre_x * nombre_y

   nombre_x = nombre_x_som
   nombre_y = nombre_y_prod

   print(f"Voici les calcules {nombre_x} et {nombre_y}")
```

```python
# Exercice 9 : calcul de la moyenne de trois notes et mention
# Objectif : utiliser les entrées utilisateur, les opérations arithmétiques et les structures conditionnelles.

note_1 = float(input("Veuillez entrer la 1ere note : "))
note_2 = float(input("Veuillez entrer la 2e note : "))
note_3 = float(input("Veuillez entrer la 3e note : "))

moy = (note_1+note_2+note_3) / 3

if  moy < 10 : 
    print(f"Votre moyenne est {moy} mention : Insuffisant !")
elif (moy > 10 and moy < 12) :
    print(f"Votre moyenne est {moy} mention : Passable !")
elif (moy > 12 and moy < 14) : 
    print(f"Votre moyenne est {moy} mention : Assez-bien !")
elif (moy > 14 and moy < 16 ) :
    print(f"Votre moyenne est {moy} mention : Bien !")
else : 
    print(f'Votre moyenne est {moy} mention : Très bien !')
```

```python
from math import sqrt

# Exercice 10 : Programme qui calcule les solutions d’une équation du second degré
# Objectif : utiliser les entrées utilisateur, les opérations arithmétiques et les structures conditionnelles.

a = int(input("Veuillez entrer le coefficient a  : "))
b = int(input("Veuillez entrer le coefficient b  : "))
c = int(input("Veuillez entrer le coefficient c  : "))

if a == 0 : 
    print("Ce n'est pas une équation du second degré ! ")
    exit()
    
caracter_discriminant = b**2 - 4*a*c

if caracter_discriminant > 0 : 
    x1 = (-b - sqrt(caracter_discriminant)) / (2 * a)
    x2 = (-b + sqrt(caracter_discriminant)) / (2 * a)

    print(f"Les soutions de l'équation  : {x1}, {x2}")

elif caracter_discriminant == 0 : 
    x = (-b) / (2 * a)
    print(f"Double solution {x}")

else : 
    print("Il n’y a pas de solution réelle")

```
