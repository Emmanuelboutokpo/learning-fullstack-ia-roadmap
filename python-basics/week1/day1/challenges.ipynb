```python
#Challenge 1 : Demander le nombre de copies de livre expédiées et calculer le coût total de gros.
#Objectif : utiliser les entrées utilisateur et les opérations arithmétiques.

nbre_copie = int(input("Combien de copie expédiez-vous ? "))

prix_public = 24.95
rabais_librairie = 0.4

prix_reduit_librairie = prix_public - (prix_public * rabais_librairie)

prix_first_expe = 3
prix_exp_sup = 0.75 * (nbre_copie-1)

prix_total_expe = prix_first_expe + prix_exp_sup

coute_total = nbre_copie * prix_reduit_librairie + prix_total_expe

print(f"Le coût total de gros de{nbre_copie} copies est {coute_total} €")
```

```python
# Challenge : Calculer l'heure d'arrivée à la maison
# Objectif : manipuler le temps (secondes, minutes, heures)

distance_1 = 1.5
rythm_1 = (5 * 60) + 10

distance_2 = 5
rythm_2 = (4 * 60) + 20

distance_3 = 1.5
rythm_3 = (5*60)+10

depart = (60 * 60 * 6) + (60 * 52)

temps_total = (distance_1 * rythm_1) + (distance_2 * rythm_2) + (distance_3 * rythm_3)

arrive = temps_total + depart

heure = int(arrive // 3600)
minute = int((arrive % 3600) // 60)
seconde = int(arrive % 60)


print(f"Vous arriverez à la maison à {heure}:{minute}:{seconde}")
 
```
