```python
# Exercice 4 : Demander l'année et calculer l'âge
# Objectif : utiliser les entrées utilisateur et les opérations arithmétiques.

DATA_NOW = 2026
annee_de_naissance = int(input('Quel est votre année de naissance ? '))
age = DATA_NOW - annee_de_naissance

print(f"Vous êtes né en {annee_de_naissance} et vous avez actuellement {age} ans !")
```

```python
# Exercice 5 : Aire et le perimetre d’un rectangle
# Objectif : apprendre à utiliser l'assignation multiple des variables pour 
# calculer l’aire et le perimetre à partir de la longueur et largeur

longueur, largeur = float(input("Quelle est la longueur du champ ? ")), float(input("Quelle est la largeur du champ ? "))

perimetre = 2* (longueur + largeur)

aire = longueur * largeur

print(f"La longueur et largeur du champ sont repectivement {longueur} et {largeur} alors le perimetre du champs est {perimetre} et l'aire du champ est {aire}")
```

```python
# Exercice 5 : Convertir des euros en dollars avec un taux fixe.
# Objectif : utiliser les entrées utilisateur et les opérations arithmétiques.

val_en_euro = float(input("Quel est votre valeur en euro ? "))

val_en_dollars = val_en_euro*1.16

print(f"La valeur en dollars de {val_en_euro} est {val_en_dollars} $")
```

```python
# Exercice 7 : Calculer le volume d'une sphère.
# Objectif : utiliser les entrées utilisateur et les opérations arithmétiques.

rayons = float(input("Quel est le rayon de sphère ?"))
volume = 4/3 * 3.14*rayons**3
print(f"Le volume d'une sphère de {rayons} est {volume}")
```
