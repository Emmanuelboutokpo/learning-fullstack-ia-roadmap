```python
# Exercice 3 : Aire d’un rectangle
# Objectif : apprendre à utiliser l'assignation multiple des variables pour 
# calculer l’aire à partir de la longueur et largeur

longueur, largeur = float(input("Quelle est la longueur du champ ? ")), float(input("Quelle est la largeur du champ ? "))

aire = longueur * largeur

print(f"L'aire du champ est {aire}")
```

```python
# Exercice 4 : Convertir des euros en dollars avec un taux fixe.
# Objectif : utiliser les entrées utilisateur et les opérations arithmétiques.

val_en_euro = float(input("Quel est votre valeur en euro ? "))

val_en_dollars = val_en_euro*1.16

print(f"La valeur en dollars de {val_en_euro} est {val_en_dollars} $")
```

```python
# Exercice 6 : Calculer le volume d'une sphère.
# Objectif : utiliser les entrées utilisateur et les opérations arithmétiques.

rayons = float(input("Quel est le rayon de sphère ?"))
volume = 4/3 * 3.14*rayons**3
print(f"Le volume d'une sphère de {rayons} est {volume}")
```

```python
# Challenge : Calculer l'heure d'arrivée à la maison
# Objectif : manipuler le temps (secondes, minutes, heures)

distance_1 = 1.5
rythm_1 = (5 * 60) + 10

distance_2 = 5
rythm_2 = (4 * 60) + 20

distance_3 = 1.5
rythm_3 = (5*60)+10

depart = (60 * 60 * 6) + (60 * 52)

temps_total = (distance_1 * rythm_1) + (distance_2 * rythm_2) + (distance_3 * rythm_3)

arrive = temps_total + depart

heure = arrive // 3600
minute = (arrive % 3600) // 60
seconde = (arrive % 60)

heure_str = str(heure).zfill(2)
minute_str = str(minute).zfill(2)
seconde_str = str(seconde).zfill(2)

print(f"Vous arriverez à la maison à {heure_str}:{minute_str}:{seconde_str}")
 
```
