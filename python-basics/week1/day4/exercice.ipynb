```python
# Exercice : Programme qui affiche le résultat de lancement d’un dé
# Objectif :   utiliser les fonctions, le hasard, les boucles, les conditions et les entrées utilisateurs
import random

compte = 0

def afficher_resultat(result) :
    print(f"Vous avez {result} !")
  
def afficher_compteur(resultat) :
    print(f"compteur : {resultat} ")

def lancer_de() :
    mon_de_lancer = random.randint(1, 6)
    if mon_de_lancer == 2 or mon_de_lancer == 5:
        return True
    else :
        return False


for n in range(0,1000) :
    verifier = lancer_de()
    if (verifier) :
        compte +=1  

afficher_compteur(compte)  

for m in range(0,10) :
    verifier = lancer_de()
    if (verifier) :
        afficher_resultat('gagné')
    else : 
        afficher_resultat('perdu')
         
```
