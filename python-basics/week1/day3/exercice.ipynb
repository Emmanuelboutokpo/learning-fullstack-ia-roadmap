```python
# Exercice 1 : Demande un nombre de depart et afficher les 10 nombres suivants avec une boucle
# Objectif : apprendre à utiliser les boucles et les entrées utilisateur.

nombre = int(input("Quel est votre nombre de depart ? "))

for nombre in range(nombre+1, nombre+11) : 
    print(nombre)
```

```python
# Exercice 2 : Calcul de la factorielle d’un nombre entier
# Objectif : utiliser les boucles, les conditions et les entrées utilisateur

nombre = int(input("Veuillez entrer un nombre : "))

# Vérification que le nombre est positif
while nombre < 0:
    nombre = int(input("Veuillez entrer un nombre positif : "))

# Cas particulier : 0! = 1
if nombre == 0:
    print(f"La factorielle de {nombre} est 1")
else:
    factorielle = 1
    for i in range(1, nombre + 1):
        factorielle *= i

    print(f"La factorielle de {nombre} est {factorielle}")
```

```python
# Exercice 4 : fonction qui renvoie la table de multiplication
# Objectif : Manipuler les fonctions, les boucles et entrer utilisateurs

def table_de_multi(nombre) : 
    if nombre > 0 :
        for i in range(0, 13) :
            print(f"{nombre} * {i} = {nombre*i}")
    else : 
        print("Le nombre doit être strictement positif !")

table_de_multi(2)
```
