```python
# Une calculatrice d'IMC (Indice de Masse Corporelle) dans la console (le terminal noir).
# Objectif : apprendre à utiliser les entrées utilisateur et les types de données.

poids = 0
taille = 0

while taille <= 0 or poids <= 0 :
    try :
      poids = float(input("Entrez votre poids en kilogrammes : "))
      taille = float(input("Entrez votre taille en mètres : "))

    except : 
        print("Erreur : Veuillez entrer des valeurs numériques valides pour le poids et la taille.")
     
imc = poids / (taille ** 2)

if imc < 18.5 :
    print(f"Votre IMC est de {imc:.2f}. Vous êtes en sous-poids.")
elif 18.5 <= imc < 24.9 :
    print(f"Votre IMC est de {imc:.2f}. Vous avez un poids normal.")
elif 25 <= imc < 29.9 :
    print(f"Votre IMC est de {imc:.2f}. Vous êtes en surpoids.")
else :
    print(f"Votre IMC est de {imc:.2f}. Vous êtes obèse.")
```
