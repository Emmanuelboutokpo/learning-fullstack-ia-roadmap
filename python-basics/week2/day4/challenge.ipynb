```python
# challenge : Mini gestionnaire de contacts poussé
# Objectif : apprendre à utiliser les listes, les dictionnaires, les fonctions, les boucles et les conditions.

contacts = []
liste_finale = []   

def saisir_contact():
    nom = input("Entrez votre nom s'il vous plaît : ")
    numero = input("Entrez votre numéro de téléphone s'il vous plaît : ")
    return  verifier_contact(nom, numero)

def verifier_contact(nom,numero):
    if len(numero) < 8:
       return affichage_message("Le numéro de téléphone doit contenir au moins 8 chiffres.")
    elif not numero.isdigit():
        return affichage_message("Le numéro de téléphone doit contenir uniquement des chiffres.")
    elif nom == "":
        return  affichage_message("Le nom ne peut pas être vide.")
    else:
        return ajouter_contact(nom, numero)

def ajouter_contact(nom, numero):
    contact = {"nom": nom, "numero": numero}
    contacts.append(contact)
    return  affichage_message("Contact ajouté avec succès.")

def verifier_contact_existant(nom, numero):
    for contact in contacts:
        if contact["nom"] == nom and contact["numero"] == numero:
            return True
    return False

def add_contact_final(nom, numero):
    if verifier_contact_existant(nom, numero):
        return affichage_message("Ce contact existe déjà dans la liste finale.")
    else:
        contact = {"nom": nom, "numero": numero}
        liste_finale.append(contact)
        return afficher_contacts("Contact ajouté à la liste finale avec succès." + str(liste_finale))

def affichage_message(message):
    print(message)
    return False

def afficher_contacts(message):
    print(message)


def menu():
    while True:
        print("\nMenu:")
        print("1. Ajouter un contact")
        print("2. Ajouter un contact à la liste finale")
        print("3. Afficher les contacts")
        print("4. Afficher la liste finale")
        print("5. Quitter")
        choix = input("Choisissez une option (1-5): ")

        if choix == '1':
            saisir_contact()
        elif choix == '2':
            nom = input("Entrez le nom du contact à ajouter à la liste finale : ")
            numero = input("Entrez le numéro du contact à ajouter à la liste finale : ")
            add_contact_final(nom, numero)
        elif choix == '3':
            afficher_contacts("Liste des contacts: " + str(contacts))
        elif choix == '4':
            afficher_contacts("Liste finale des contacts: " + str(liste_finale))
        elif choix == '5':
            print("Au revoir!")
            break
        else:
            print("Option invalide. Veuillez réessayer.")

if __name__ == "__main__":
    menu()
```
